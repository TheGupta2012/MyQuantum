# Simulating the law of large numbers for a die roll

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
n = int(input("Enter the number of trials:"))
trials =[]
values =[]
for i in range(n):
    val =0
    for k in range(i):
        val += random.randint(1,6)
    val = val/(i+1)
    trials.append(i+1)
    values.append(val)
plt.xlabel("Number of trials",fontsize=15)
plt.ylabel("Mean of the Value rolled")
sns.lineplot(trials,values)
# plt.lab("The random throws"
sns.lineplot(trials,[3.5 for i in range(n)])
plt.legend(['The random throws','Expected value'])
plt.savefig("D:/IBM-Quantum/Course Work/Week5 - Probabibility/law_of_large_numbers.png",dpi=200)
```
