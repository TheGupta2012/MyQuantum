## Some math 
- Add |+> and |-> in numpy 
- (|0>+|1>)/sqrt(2) , (|0>-|1>)/sqrt(2)
- First create basis states

```python
A = np.array([1,0])
B = np.array([0,1])
```

```python
C = (A+B)/np.sqrt(2)
```

```python
D = (A-B)/np.sqrt(2)
```

```python
print("C:",C)
print("D:",D)
```

## Transposing 
- arrayname.T returns the transposed state of your vector
- .T is an ATTRIBUTE 
- .conj() is a function
- **[[]]** is a way of specifying that we ACTUALLY have a row or a column vector

```python
#practice 
# make A = 1/sqrt(2)(1+j)
A = np.array([[1,1j]]/np.sqrt(2))
```

```python
print("A:",A)
```

```python
print("Transpose:",A.T)
```

```python
print("Conjugate Transpose:",A.conj())
```

## How to multiply two matrices?
- A @ B is the way to multiply the matrices A and B 
- Let's do a problem -
How is the library dealing with the non-compatibility of state and Z in the first multiplication?

```python
Image('image2.JPG')
```

```python
state = np.array([0,1])# row , that is why answer as row only
Z = np.array([[1,0],[0,-1]])
H = np.array([[1,1],[1,-1]])/ np.sqrt(2)
```

```python
print("State :",state)
```

```python
print("Applying Z:",Z @ state)
```

```python
print("Applying H:",H @ Z @ state)
```
