## IBM Quantum Experience 
- It is a cloud based quantum computing platform
- Users can create quantum circuits or in code using qiskit
- Circuits either run on QASM( Quantum Assembly Language Simulator) or hardware 
- When a circuit is run, it joins a **queue**

### Difference in QASM and Hardware
- **Hardware** here means that the code is actually being executed on the actual quantum computer
- **SIMULATOR** - It is a classical computer 'pretending' to be a quantum computer. It just knows how to do the math.


### Queue is just a FIFO job queue

### QASM is required for testing during the labs (queue problems)

# Practice
- The measurement done in the ibmQ is done in the Z basis

```python
from qiskit.tools.visualization import plot_histogram,plot_bloch_multivector
```

```python
Q = QuantumCircuit(1,1)
Q.h(0)
Q.draw(output='mpl')
```

```python
state = execute(Q,backend = Aer.get_backend('statevector_simulator')).result().get_statevector()
print(state)
```

```python
plot_bloch_multivector(state)
```

### NOTE
- The quantum computer doesn't always have an operation mapped to your gates
- **TRANSPILATION** is done for the circuit to be *runnable on the actual hardware*

```python
q2 = QuantumCircuit(2)
q2.h(0)
q2.cx(0,1)
q2.draw(output='mpl')
```

```python
q2.measure_all()
```

```python
counts = execute(q2,backend = Aer.get_backend('qasm_simulator'),shots=2000).result().get_counts()
print(counts)
```
