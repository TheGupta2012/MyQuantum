# Handling noise in Quantum
- Here, we aspire to correct the noisy results we see in our circuits 
- Till now we built the circuits in <b> qiskit terra and simulated in qiskit Aer </b>
- Now we are going to execute the circuits on <b>qiskit ignis</b>
- Well why <b>ignis</b> ? to characterize errors 
- There are various kinds of errors in Quantum
- In this notebook, focus is on <b> MEASUREMENT ERROR MITIGATION</b>


## IDEA 
- Suppose we create a $|11\rangle$ state.
- On real hardware we get some shots of the other states that are 00,01 and 10.
- What if we knew the chances of seeing the wrong answer for each of these initial states
- Maybe we could use that to make the chances of measuring the <b> required state </b> with a higher probability.


```python
# building a simple circuit 
from qiskit import *
print(qiskit.__qiskit_version__)
%matplotlib inline 
Q = QuantumCircuit(3,3)
Q.h(0)
Q.cx(0,1)
Q.cx(1,2)
Q.measure([0,1,2],[0,1,2])
Q.draw(output='mpl')
```

## Why triple entanglement?
- Well all initial states are 0,0,0
- In the first qubit, you apply a hadamard 
- Suppose you get a zero in the q0( 50 - 50probability )
- Then since q1 was initially 0, it <b> stays in 0 state</b> as Cx gate doesn't do anything.
- Now, q1 is 0 so Cx from 1->2 doesn't change the state of q2 and thus we get the -
    - $|000\rangle$ state
- Similar explanation can be derived in the case when q0 was initially 1.

```python
from qiskit.tools.visualization import plot_histogram 
sim = Aer.get_backend('qasm_simulator')
result = execute(Q,backend=sim,shots = 1024).result()
count = result.get_counts()
plot_histogram(count)
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider(hub='ibm-q')
device = provider.get_backend('ibmqx2')
```

```python
job = execute(Q,backend = device, shots = 1024)
print(job.job_id())
from qiskit.tools.monitor import job_monitor 
job_monitor(job)
```

```python
dev_result = job.result()
plot_histogram(dev_result.get_counts())
```

## So there is noise we see...

```python
Q.qregs[0]
```

```python
from qiskit.ignis.mitigation.measurement import (complete_meas_cal,CompleteMeasFitter)
```

```python
calib_circ,state_labels = complete_meas_cal(qr = Q.qregs[0],circlabel='meas_calib')
```

- The above line gives us a series of calibration circuits to run corresponding to each state that we have in our quantum circuit

```python
calib_circ[3].draw(output='mpl') # 011
```

```python
len(calib_circ) # because 2^3 is 8 
```

```python
# trying to execute the circuit on a quantum device.
cal_job = execute(
                calib_circ,
                backend = device,
                shots = 1024,
                optimization_level=0
        )
print(cal_job.job_id())
job_monitor(cal_job)
cal_results = job.result()
```

```python
plot_histogram(cal_results.get_counts())
```

```python
print("Normal results :",dev_result.get_counts(),"\nCalibration circuit counts:",cal_results.get_counts())
```

## Ignis workflow
- We generate a series of circuits instead of just one circuit
- Run them on hardware
- Take the results back from the hardware and try to understand what parameters for future error mitigation come out of these results( whatever "parameters" might mean right now)
- What we are trying to do is - <b>once</b> ,run an n-qubit circuit.
- Then, see what the errors we got in our measurements on that Quantum Computer 
- <b>Fit that output to a filter</b> for error mitigation of an n-qubit system
- After we got the filter for the n-qubit system, we run another n-qubit system on the hardware
- Then <b>we used the filter we built ONCE </b>using the previous circuit and then whatever "parameters" our filter learnt in the first circuit, it applied those to **CORRECT THE OUTPUT OF THIS CIRCUIT CORRESPONDING TO THE LEARNT PARAMETERS.**

It did not just correct the errors pertaining to the highest probability results but according to the "parameters" it learned for correcting a type of the n-qubit circuit that was passed to the Quantum Hardware.
