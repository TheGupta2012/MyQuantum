## BASICS OF CIRCUITS
- Whether we are using qubits or bits, we need to manipulate them in order to turn the inputs we have into the outputs we need.
- Circuit diagrams are a useful way to represent the process and the programs containing very few qubits or bits 
### Three jobs required in quantum
- Encode the input( initially are <b> one </b>
- Do some computation on your input
- Finally extract some output out of it ( done through <b>measuring</b> )




```python
#example 
q1 = QuantumCircuit(7,7)
for i in range(7):
    q1.measure(i,i)
%matplotlib inline 
q1.draw(output = 'mpl')
```

```python
count = execute(q1,backend=Aer.get_backend('qasm_simulator'),shots=200).result().get_counts()
# shots are present because quantum computers may have randomness in them
# this case gets us all results with 100% certainity as we actually did nothing on our qubits 
plot_histogram(count)
```

- <b> Processing input and Measuring outputs</b>
    - According to the rules of basic binary addition, :
        -<b>$$ 0 + 0 = 0 , 1 + 0, 0 + 1 = 1 and 1 + 1 = 0 \ with \ carry \ 1$$</b>
    - The four basic sums are the above only and we use a half adder to implement them.
    - Addition 
        - So what we want to do is to have the sum as 1 whenever the two qubits are different
        - The carry should only be 1 when both of them are 1. 
        
### Sum - > XOR ( use double CNOT to implement)
### Carry - > Toffoli

```python
# using circuit composition 
# original circuit 
q = QuantumCircuit(4,2)
q.h([0,1])
q.barrier()
# sum circuit
sumq = QuantumCircuit(3)
sumq.cx(0,2)
sumq.cx(1,2)
# carry circuit 
carry = QuantumCircuit(3)
carry.ccx(0,1,2)
```

```python
# composing circuit 
q = q.compose(sumq,[0,1,2]) # list is the qubits we want to attach sum on
q.barrier()
q = q.compose(carry,[0,1,3]) 
q.barrier()
q.measure(2,0)
q.measure(3,1)
print("My composed half adder circuit!")
q.draw(output = 'mpl')
```

- Here, we have actually passed in a superposition of all the four states 00,01,10 and 11
- Note that for 01 and 10 the sum is always 1 and carry is always 0 and thus it appears 50% of the time
- Rest of the 2 are appearing 25% of the time only

```python
counts = execute(q,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```
