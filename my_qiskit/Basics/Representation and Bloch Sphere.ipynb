- For qubits, the restriction of always being in a well defined state is lifted.

**Whether we get a 0 or a 1 from a qubit only needs to be well-defined when a measurement is made to extract an output**

**At all other times, its state will be something more complex than can be captured by a simple binary value.**
 $$ |q_{0}\rangle = \frac{1}{\sqrt{2}}|0\rangle + \frac{i}{\sqrt{2}}|1\rangle$$
|q0âŸ© is called the qubit's statevector, it <i>tells us everything we could possibly know about this qubit.</i> It is described by a linear combination of the two basis vectors of our system

```python
# making different - different qubits
q = QuantumCircuit(1)
# these always start in the |0> state 
state = np.array([( 1 + 2j), (1-2j)])
state = state/(sqrt(10))
q.initialize(state,0)
q.draw(output= 'mpl')
```

```python
q = QuantumCircuit(1)
s = np.array([1+4j , 3-3j])
s = s/(np.linalg.norm(s))
q.initialize(s,0)
q = transpile(q,backend = Aer.get_backend('qasm_simulator'))
q.draw('mpl')
```

```python
plot_bloch_multivector(execute(q,backend= Aer.get_backend("statevector_simulator")).result().get_statevector())
```

## Properties
- The states are always always always NORMALISED.
- Means the sum of amplitudes squared must be equal to 1
 $$|\alpha|^{2}+|\beta|^{2}\ must\ be\ 1$$
- <b>Global phase factor</b> does not affect the states of our qubits, they do not change the expectation values and neither do they change the measurement probabilities of our qubits 
- Also, we are <b>not restricted</b> to measuring the state of our qubits in the |0> and |1> basis. Any basis |x> may be chosen for the measurements of our qubits.


```python
# vector with 1/3 probability of |0> 
q = QuantumCircuit(1)
state = [1/sqrt(2),-1j/sqrt(2)]
q.initialize(state,0)
display(q.draw(output = 'mpl')) 
plot_bloch_multivector(execute(q,backend = Aer.get_backend('statevector_simulator')).result().get_statevector())
```

## Simulating the collapse of our qubit
- The general state of the qubit on the Bloch Sphere is - 
 $$|\Psi \rangle = cos\frac{\theta}{2}|0 \rangle + e^{i\phi} sin\frac{\theta}{2} |1\rangle$$
 - Here, $\phi$ is the rotation along the latitude of the sphere, or the circular rotation and $\theta$ is the angle that the state makes with the positive **Z** axis

```python
q = QuantumCircuit(1,1)
q.h(0)
print("We have the qubit in equal superposition ")
print("Horizontal turn is still zero")
print("Vertical turn is pi/2")
state_v = execute(q,backend= Aer.get_backend("statevector_simulator")).result().get_statevector()
print(state_v)
plot_bloch_multivector(state_v)
```

```python
q.measure(0,0)
print("We have collapsed our state")
```

```python
state_v = execute(q,backend= Aer.get_backend("statevector_simulator")).result().get_statevector()
print(state_v)
print("Each subsequent measurement gives the same outcome")
plot_bloch_multivector(state_v)
```

## Plotting states
- Note, **qasm_simulator** has no *get_statevector* method as it tries to simulate a real quantum computer. This means, when we want the statevector of a quantum state, we need to use **statevector_simulator**

```python
q = QuantumCircuit(1)
state_v = execute(q,backend= Aer.get_backend("statevector_simulator")).result().get_statevector()
plot_bloch_multivector(state_v)
```

```python
q.x(0)
state_v = execute(q,backend= Aer.get_backend("statevector_simulator")).result().get_statevector()
plot_bloch_multivector(state_v)
```

```python
q.x(0) # inverse of x is x 
# equal superposition
q.h(0)
state_v = execute(q,backend= Aer.get_backend("statevector_simulator")).result().get_statevector()
plot_bloch_multivector(state_v)
```
